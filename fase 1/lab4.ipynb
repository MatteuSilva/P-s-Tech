```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.figure as figure
dados = pd.read_csv('A150850189_28_143_208.csv', encoding='ISO-8859-1', skiprows=3, sep=';', thousands='.', decimal=',', skipfooter=12)
print(dados)
```

```python
def is_numeric_column(series):
    return pd.to_numeric(series, errors='coerce').notna().all()

numeric_cols = dados.columns[dados.apply(is_numeric_column, axis=0)]
dados_numericos = dados[numeric_cols]

colunas_usaveis = dados_numericos.mean().index.tolist()
print(colunas_usaveis)
```

```python
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis[:5]
```

```python
dados_usaveis = dados[colunas_usaveis]
dados_usaveis.head()
```

```python
dados_usaveis = dados_usaveis.set_index("Unidade da Federação")
dados_usaveis.head()
```

```python
dados_usaveis["Total"] = dados_usaveis.sum(axis = 1)
dados_usaveis = pd.DataFrame(dados_usaveis)
dados_usaveis = dados_usaveis.drop("2007/Ago", axis=1)
dados_usaveis = dados_usaveis.drop("2007/Set", axis=1)
dados_usaveis = dados_usaveis.drop("2007/Out", axis=1)
dados_usaveis = dados_usaveis.drop("2007/Nov", axis=1)
dados_usaveis = dados_usaveis.drop("2007/Dez", axis=1)
dados_usaveis.head()
```

```python
ordenados_por_total = dados_usaveis.head(7).sort_values(by='Total', ascending=False)
ordenados_por_total = ordenados_por_total / 1_000_000
axis = ordenados_por_total.head(5).T.plot(figsize=(12,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.0f}"))
```
