```python
import pandas as pd
import matplotlib
dados = pd.read_csv('sih_cnv_qiuf194843177_221_129_62.csv', encoding='ISO-8859-1', skiprows=3, sep=';', thousands='.', decimal=',', skipfooter=12)
print(dados)
```

```python
dados['2008/Jan'].mean()
```

```python
dados.plot(x="Região/Unidade da Federação", y="2008/Ago")
```

```python
import matplotlib.ticker as ticker
import matplotlib.pyplot as plt

axis = dados.plot(x="Região/Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))

plt.title("Valor por unidade da federação")
plt.show()
```
